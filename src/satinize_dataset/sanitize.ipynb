```python
from turtle import mode
import pandas as pd

# Load the CSV file into a DataFrame
emdat_disasters_df = pd.read_csv("original_datasets/emdat_disasters.csv", encoding="utf-8")
emdat_disasters_df.info()
```

```python
def removeColumns(df):
    return df.drop(
        columns=[
            "Historic",
            "Classification Key",
            "External IDs",
            "Entry Date",
            "Last Update",
            "Admin Units",
            "Disaster Subtype",
            "Origin",
            "River Basin",
            "Location",
            "End Year",
            "End Month",
            "End Day",
            "OFDA/BHA Response",
            "Appeal",
            "Declaration",
            "CPI",
            "Reconstruction Costs ('000 US$)",
            "Reconstruction Costs, Adjusted ('000 US$)",
            "Associated Types",
            "AID Contribution ('000 US$)",
            "Magnitude",
            "Magnitude Scale",
            "Latitude",
            "Longitude",
            "Start Month",
            "Start Day",
            "No. Injured",
            "No. Affected",
            "No. Homeless",
            "Total Affected",
            "Insured Damage ('000 US$)",
            "Insured Damage, Adjusted ('000 US$)",
            "Total Damage ('000 US$)",
            "Event Name",
            "Total Damage, Adjusted ('000 US$)",
        ],
        axis=1,
    )
emdat_disasters_df = emdat_disasters_df.rename(columns={"DisNo.": "DisNo", "Disaster Group": "DisasterGroup", "Disaster Subgroup": "DisasterSubgroup", "Disaster Type": "DisasterType", "Start Year": "StartYear", "Total Deaths": "TotalDeaths"})
emdat_disasters_df = emdat_disasters_df.pipe(removeColumns)

emdat_disasters_df['TotalDeaths'] = emdat_disasters_df['TotalDeaths'].fillna(0)

emdat_disasters = emdat_disasters_df
emdat_disasters
```

```python
country_data = pd.read_csv("original_datasets/world-data-2023.csv", encoding="utf-8")

# Correct the names of countries in both datasets that don't match
emdat_disasters_df.loc[emdat_disasters_df['Country']=='Türkiye', 'Country'] = 'Turkey'
emdat_disasters_df.loc[emdat_disasters_df['Country']=='United States of America', 'Country'] = 'United States'
emdat_disasters_df.loc[emdat_disasters_df['Country']=='Russian Federation', 'Country'] = 'Russia'
emdat_disasters_df.loc[emdat_disasters_df['Country']=='Czechia', 'Country'] = 'Czech Republic'
emdat_disasters_df.loc[emdat_disasters_df['Country']=='Netherlands (Kingdom of the)', 'Country'] = 'Netherlands'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Democratic People's Republic of Korea", 'Country'] = 'North Korea'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Republic of Korea", 'Country'] = 'South Korea'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Congo", 'Country'] = 'Republic of the Congo'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Iran (Islamic Republic of)", 'Country'] = 'Iran'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="United Republic of Tanzania", 'Country'] = 'Tanzania'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Viet Nam", 'Country'] = 'Vietnam'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Syrian Arab Republic", 'Country'] = 'Syria'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Cabo Verde", 'Country'] = 'Cape Verde'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Venezuela (Bolivarian Republic of)", 'Country'] = 'Venezuela'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Timor-Leste", 'Country'] = 'East Timor'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Bolivia (Plurinational State of)", 'Country'] = 'Bolivia'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Republic of Moldova", 'Country'] = 'Moldova'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Ireland", 'Country'] = 'Republic of Ireland'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="United Kingdom of Great Britain and Northern Ireland", 'Country'] = 'United Kingdom'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Micronesia (Federated States of)", 'Country'] = 'Federated States of Micronesia'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="State of Palestine", 'Country'] = 'Palestinian National Authority'
emdat_disasters_df.loc[emdat_disasters_df['Country']=="Bahamas", 'Country'] = 'The Bahamas'

# Discard countries with no match on the two datasets (doing so only discards about 362 events which is a small percentage of the dataset (~2.3%))
non_overlapping_countries = [country for country in country_data['Country'].to_list() if country not in list(emdat_disasters_df['Country'].unique())]
for country in non_overlapping_countries:
    country_data = country_data.drop(country_data[(country_data['Country'] == country)].index)

non_overlapping_countries = [country for country in list(emdat_disasters_df['Country'].unique()) if country not in country_data['Country'].to_list()]
for country in non_overlapping_countries:
    emdat_disasters_df = emdat_disasters_df.drop(emdat_disasters_df[(emdat_disasters_df['Country'] == country)].index)

# keep only desired columns
columns_to_keep = ['Country', 'Land Area(Km2)']
country_data = country_data[columns_to_keep]
country_data['Land Area(Km2)'] = country_data['Land Area(Km2)'].str.replace(',', '').astype('Int64')
country_data.rename(columns={"Land Area(Km2)": "land_area"}, inplace=True)

# Remove palestine from the dataset
country_data = country_data[country_data['Country'] != 'Palestinian National Authority']

country_data
```

```python
# create a new DataFrame with the number of disasters and deaths per year per contry, region and subregion and save it to a new CSV file
# Group by Year, Country, Region, and Subregion and calculate the number of disasters and total deaths
disasters_per_country = emdat_disasters_df.groupby(['StartYear', 'Country', 'ISO', 'Region']).agg(
    num_disasters=('DisNo', 'count'),
    total_deaths=('TotalDeaths', 'sum')
).reset_index()

# Merge the new DataFrame with the country_data DataFrame to add the Land Area(Km2) column
disasters_per_country = pd.merge(disasters_per_country, country_data, on='Country')

#Calculating the number of disasters per area
disasters_per_country['disasters_per_area'] = (disasters_per_country['num_disasters'] / disasters_per_country['land_area']) * 10**7


# Save the new DataFrame to a CSV file
disasters_per_country.to_csv(
    "pre-processing/disasters_per_country.csv", index=False, mode="w", encoding="utf-8"
)

disasters_per_country
```

```python
# Group by Year, Country, and DisasterSubgroup and calculate the frequency of each disaster subgroup
disaster_subgroup_frequency = emdat_disasters_df.groupby(['StartYear', 'Country', 'ISO', 'DisasterGroup', 'DisasterSubgroup']).size().reset_index(name='frequency')

# Calculate the number of deaths for each disaster subgroup in each year and country
deaths_per_subgroup = emdat_disasters_df.groupby(['StartYear', 'Country', 'DisasterGroup', 'DisasterSubgroup'])['TotalDeaths'].sum().reset_index(name='total_deaths')

# Merge the deaths data with the frequency data
disaster_subgroup_frequency = pd.merge(disaster_subgroup_frequency, deaths_per_subgroup, on=['StartYear', 'Country', 'DisasterGroup', 'DisasterSubgroup'])

# remove extra terrestrial
disaster_subgroup_frequency = disaster_subgroup_frequency[disaster_subgroup_frequency['DisasterSubgroup'] != 'Extra-terrestrial']

# Save the new DataFrame to a CSV file
disaster_subgroup_frequency.to_csv('pre-processing/disaster_subgroup_frequency.csv', index=False, mode='w', encoding='utf-8')

disaster_subgroup_frequency
```
